```python
from datasets import load_dataset
from transformers import pipeline
```

```python
# The following is an example
# loading the dataset
xsum_dataset = load_dataset(
    "xsum",
    version="1.2.0",
    cache_dir='/Documents/Huggin_Face/data'
)  # Note: We specify cache_dir to use predownloaded data.
xsum_dataset
# The printed representation of this object shows the `num_rows`
# of each dataset split.
```

```python
xsum_sample = xsum_dataset["train"].select(range(10))

display(xsum_sample.to_pandas())
```

```python
summarizer = pipeline(
    task="summarization",
    model="t5-small",
    min_length=20,
    max_length=40,
    truncation=True,
    model_kwargs={"cache_dir": '/Documents/Huggin_Face/'},
)  # Note: We specify cache_dir to use predownloaded models.
```

```python
summarizer(xsum_sample["document"][0])
```

```python
import tkinter as tk

def process_input():
    user_input = input_entry.get()  # Get text from the input field
    #Summarizer
    summary = summarizer(user_input, max_length=150, min_length=30, do_sample=False)[0]['summary_text']

    bullet_points = summary.split(". ")
    for point in bullet_points:

        print(f"- {point}")
    # Perform some processing on the input (e.g., reversing it)
    processed_output = summary

    # Display the processed output in the output label
    output_text.config(text=f"Processed Output: {processed_output}")

root = tk.Tk()
root.title("Text Input and Output GUI")

# Create and place an input field
input_entry = tk.Entry(root)
input_entry.pack()

# Create and place a button to process input
process_button = tk.Button(root, text="Process Input", command=process_input)
process_button.pack()

# Create and place a label for displaying output
output_text = tk.Label(root, text="")
output_text.pack()

root.mainloop()
```
